```python
def mad_libs():
    story = "Roses are [adjective], violets are [color], sugar is [adjective], and so are you!"
    placeholders = ['[adjective]', '[color]']
    user_inputs = []

    for placeholder in placeholders:
        user_input = input(f"Enter an {placeholder.strip('[]')}: ")
        user_inputs.append(user_input)

    completed_story = story.replace('[adjective]', user_inputs[0]).replace('[color]', user_inputs[1])
    print("\nYour Mad Libs story:\n")
    print(completed_story)

mad_libs()
```
