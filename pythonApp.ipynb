```python
import streamlit as st

# Page Configuration
st.set_page_config(page_title="My Website", page_icon="🌍", layout="wide")

# Title
st.title("Welcome to My Streamlit Website 🚀")

# Sidebar Navigation
st.sidebar.header("Navigation")
page = st.sidebar.radio("Go to", ["Home", "About", "Contact"])

# Page Content
if page == "Home":
    st.header("Home Page 🏠")
    st.write("This is the homepage of my website.")
    st.image("https://source.unsplash.com/800x400/?nature,water")

elif page == "About":
    st.header("About Page 📜")
    st.write("This website is built using **Streamlit** in Python!")

elif page == "Contact":
    st.header("Contact Me 📞")
    st.write("Email: myemail@example.com")
    st.write("GitHub: [My GitHub](https://github.com/)")

# Footer
st.markdown("---")
st.markdown("© 2025 My Website | Built with ❤️ using Streamlit")
```
